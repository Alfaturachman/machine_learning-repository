## Pendahuluan
Decision Tree adalah algoritma pembelajaran terbimbing yang digunakan untuk masalah klasifikasi dan regresi. Algoritma ini membagi data menjadi sub-himpunan berdasarkan nilai fitur yang paling signifikan, membentuk struktur seperti pohon. Dalam proyek ini, kita menggunakan Decision Tree untuk mengklasifikasikan spesies bunga Iris (Setosa, Versicolor, Virginica) berdasarkan fitur morfologisnya.

### Alur Kerja (Pipeline):
1. **Data Acquisition**: Mengambil dataset Iris dari pustaka Scikit-Learn.
2. **Data Understanding**: Memahami fitur dan target kelas pada dataset.
3. **Modeling**: Membangun dan melatih model DecisionTreeClassifier.
4. **Prediction**: Melakukan prediksi pada data sampel atau data pengujian.
5. **Evaluation**: Mengukur performa model menggunakan Confusion Matrix dan Classification Report.
6. **Visualization**: Menampilkan struktur pohon keputusan yang telah dihasilkan.

## 1. Data Acquisition & Library Imports
Langkah pertama adalah mengimpor pustaka yang diperlukan dan memuat dataset Iris.

```python
#Import Library Scikit Learn untuk Decision Tree, Dataset & Matplotlib
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
#Load Dataset Iris dari library scikit learn
iris = datasets.load_iris()
features = iris['data']
target = iris['target']
```

## 2. Data Understanding
Eksplorasi data untuk melihat fitur-fitur (sepal length, sepal width, petal length, petal width) dan target label kelas.

```python
print(features)
print(target)
```

```python
iris['feature_names']
```

```python
iris['target_names']
```

```python
iris
```

## 3. Model Training
Membangun model Decision Tree dengan parameter default dan melatihnya menggunakan seluruh fitur dan target yang tersedia.

```python
# Membuat Objek Model Decision Tree
decisiontree = DecisionTreeClassifier(random_state=0, max_depth=None, 
                                      min_samples_split=2, min_samples_leaf=1,
                                      min_weight_fraction_leaf=0,
                                      max_leaf_nodes=None, 
                                      min_impurity_decrease=0)
```

```python
#Mentraining Model Decision Tree
model = decisiontree.fit(features, target)
```

## 4. Prediction
Melakukan prediksi terhadap data yang sama untuk melihat hasil klasifikasi, serta melakukan uji coba pada sampel observasi baru.

```python
test = model.predict(features)
```

```python
print(target)
print(test)
```

```python
#Mengambil sampel observasi dan membuat prediksi
#Sampel berupa data dimensi kelopak
#Fungsi predict() => memeriksa kelas yang dimilikinya
#Fungsi predict_proba() => memeriksa probabilitas kelas dari prediksi tersebut
observation = [[8, 5.5, 2, 8]]
model.predict(observation)
model.predict_proba(observation)

print(observation)
print(model.predict(observation))
# print(model.predict_proba(observation))
```

## 5. Performance Evaluation
Mengevaluasi model menggunakan Confusion Matrix dan Classification Report untuk melihat metrik akurasi prediksi.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target, test)
print(cm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(8,5))
sns.heatmap(confusion_matrix(target, test), annot=True, fmt=".0f", ax=ax)
plt.xlabel("test")
plt.ylabel("target")
plt.show()
```

```python
from sklearn.metrics import classification_report
print (classification_report(target, test))
```

## 6. Visualization
Membuat representasi visual dari pohon keputusan untuk memahami logika percabangan yang digunakan oleh model.

```python
# Membuat grafik visualisasi Decision Tree menggunakan plot_tree
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
tree.plot_tree(decisiontree, 
               feature_names=iris['feature_names'], 
               class_names=iris['target_names'], 
               filled=True, 
               rounded=True)

# Menyimpan hasil visualisasi ke file gambar
plt.savefig('iris.png')
plt.show()
```

## Kesimpulan
Algoritma Decision Tree sangat efektif untuk dataset Iris karena mampu memisahkan kelas-kelas dengan aturan logika yang sederhana dan mudah diinterpretasikan. Visualisasi pohon menunjukkan fitur mana (seperti petal width) yang paling dominan dalam menentukan jenis bunga Iris.
